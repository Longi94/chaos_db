```python
%matplotlib notebook
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt


def read_sqlite(file):
    conn = sqlite3.connect(file)
    df = pd.read_sql_query('SELECT * FROM result', conn)
    conn.close()
    return df
```

```python
df_0_1 = read_sqlite('results/sqlite_tpch1_flip_heap_0.1_results.sqlite')
df_0_1['rate'] = 0.1
df_0_1
```

```python
df_0_31 = read_sqlite('results/sqlite_tpch1_flip_heap_0.31_results.sqlite')
df_0_31['rate'] = 0.31
df_0_31
```

```python
df_1 = read_sqlite('results/sqlite_tpch1_flip_heap_1_results.sqlite')
df_1['rate'] = 1.0
df_1
```

```python
df_3_162 = read_sqlite('results/sqlite_tpch1_flip_heap_3.162_results.sqlite')
df_3_162['rate'] = 3.162
df_3_162
```

```python
df_10 = read_sqlite('results/sqlite_tpch1_flip_heap_10_results.sqlite')
df_10['rate'] = 10
df_10
```

```python
df_31_622 = read_sqlite('results/sqlite_tpch1_flip_heap_31.622_results.sqlite')
df_31_622['rate'] = 31.622
df_31_622
```

```python
df_100 = read_sqlite('results/sqlite_tpch1_flip_heap_100_results.sqlite')
df_100['rate'] = 100
df_100
```

```python
results_counts = {}

merged_df = pd.concat([
    df_0_1,
    df_0_31,
    df_1,
    df_3_162,
    df_10,
    df_31_622,
    df_100
])

df = merged_df.groupby(['result', 'rate']).size().reset_index(name='counts')

for result in df['result'].unique():
    for rate in df['rate'].unique():
        if len(df[(df['result'] == result) & (df['rate'] == rate)]) == 0:
            df = df.append({
                'result': result,
                'rate': rate,
                'counts': 0
            }, ignore_index=True)

df['result'] = df['result'].astype(int)
df['counts'] = df['counts'].astype(int)
df.sort_values(['result', 'rate'], inplace=True)

df
```

```python
fig = plt.figure()
fig.suptitle('Results')
plt.ylabel('Result count')
plt.xlabel('Fault rate')
plt.xscale('log')
rates = df[df['result'] == 0]['rate']
plt.plot(rates, df[df['result'] == 0]['counts'], 'b-', label='ok')
plt.plot(rates, df[df['result'] == 1]['counts'], 'r-', label='incorrect')
plt.plot(rates, df[df['result'] == 2]['counts'], 'g-', label='crash')
plt.plot(rates, df[df['result'] == 3]['counts'], 'y-', label='abnormal')
plt.plot(rates, df[df['result'] == 4]['counts'], '-', color='gray', label='timout')
plt.legend()
```

```python
fig = plt.figure()

fig.suptitle('Stacked')
plt.ylabel('Result count')
plt.xlabel('Fault rate')
plt.xscale('log')
plt.stackplot(rates, 
              df[df['result'] == 0]['counts'],
              df[df['result'] == 1]['counts'],
              df[df['result'] == 2]['counts'],
              df[df['result'] == 3]['counts'],
              df[df['result'] == 4]['counts'],
              baseline='zero')
```

```python
fig = plt.figure()

fig.suptitle('Crash runtime')

df_2 = merged_df[merged_df['result'] == 1]
data_2 = [
    list(df_2[df_2['rate'] == 0.1]['runtime']),
    list(df_2[df_2['rate'] == 0.31]['runtime']),
    list(df_2[df_2['rate'] == 1]['runtime']),
    list(df_2[df_2['rate'] == 3.162]['runtime']),
    list(df_2[df_2['rate'] == 10]['runtime']),
    list(df_2[df_2['rate'] == 31.622]['runtime']),
    list(df_2[df_2['rate'] == 100]['runtime']),
]

plt.boxplot(data_2, showfliers=False)
```
